```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 
data = {
    'Subject': ['Math', 'Science', 'Language', 'Social Studies'],
    'Easy': [40, 50, 70, 30],
    'Medium': [30, 35, 20, 50],
    'Hard': [30, 15, 10, 20]
}

#  DataFrame
df = pd.DataFrame(data)

# 
plt.figure(figsize=(8, 6))
df.set_index('Subject').plot(kind='bar', stacked=True, ax=plt.gca())
plt.title('Task Difficulty Distribution by Subject', fontsize=14)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Number of Tasks', fontsize=12)
plt.legend(title='Difficulty', fontsize=10)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 
days = pd.date_range(start="2024-01-01", periods=30)
daily_tasks = np.random.poisson(lam=50, size=30)  # 50

# 
plt.figure(figsize=(10, 6))
plt.plot(days, daily_tasks, marker='o', linestyle='-', color='b', label='Tasks Completed')
plt.title('Daily Task Completion Trend', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Number of Tasks Completed', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()

# 
total_tasks = df[['Easy', 'Medium', 'Hard']].sum()

plt.figure(figsize=(6, 6))
plt.pie(
    total_tasks,
    labels=total_tasks.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette("pastel")[0:3]
)
plt.title('Overall Task Difficulty Proportion', fontsize=14)
plt.tight_layout()
plt.show()
```
