```python
from bs4 import BeautifulSoup
import requests


# 發送 HTTP GET 請求
url = 'https://www.ettoday.net/news/20170308/880380.htm'
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 '
                  '(KHTML, like Gecko) Chrome/88.0.4324.146 Safari/537.36'
}
response = requests.get(url, headers=headers)
html = response.content


# 將 HTML 內容解析為 BeautifulSoup 對象
soup = BeautifulSoup(html, "html.parser")


# 移除 <script> 和 <style> 標籤
for script in soup(["script", "style"]):
    script.extract()


# 提取純文字內容
text = soup.get_text()


# 清理文字內容
# 將文字分成多行並移除每行首尾的空白
lines = (line.strip() for line in text.splitlines())
# 將多段文字進一步分割為單個短語
chunks = (phrase.strip() for line in lines for phrase in line.split(" "))
# 移除空白行，重新組合成一個乾淨的文字塊
texts = '\n'.join(chunk for chunk in chunks if chunk)


# 輸出清理後的文字
print(texts)
```

```python
import requests
from bs4 import BeautifulSoup

# 設定目標網址
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'

# 需要設定 cookies 通過 PTT 的 18 歲警告
cookies = {'over18': '1'}

# 發送 HTTP GET 請求
response = requests.get(url, cookies=cookies)
html = response.content

# 使用 BeautifulSoup 解析 HTML
soup = BeautifulSoup(html, "html.parser")

# 找到所有文章的標題
titles = soup.find_all('div', class_='title')

# 提取並打印標題
for i, title in enumerate(titles, start=1):
  if title.a:  # 確保標題內有超連結
      print(f"{i}. {title.a.text.strip()}")
```

```python
import requests
url = 'https://invoice.etax.nat.gov.tw/index.html'
web = requests.get(url) # 取得網頁內容
web.encoding='utf-8'
 # 因為該網頁編碼為utf-8，加
from bs4 import BeautifulSoup
soup = BeautifulSoup(web.text, "html.parser")
 # 轉換成標簽樹
td = soup.select('.container-fluid')[0].select('.etw-tbiggest')
#取出中獎號碼的位置
ns = td[0].getText()  # 特別獎
n1 = td[1].getText()  # 特獎
# 頭獎，因為存入串列會出現/n 換行符，使用[-8:] 取出最後八碼
n2 = [td[2].getText()[-8:], td[3].getText()[-8:], td[4].getText()[-8:]]
print(ns)
print(n1)
print(n2)
```

```python
import requests
url = 'https://rate.bot.com.tw/xrt/flcsv/0/day'   # 牌告匯率 CSV 網址
rate = requests.get(url)   # 爬取網址內容
rate.encoding = 'utf-8'    # 調整回應訊息編碼為 utf-8，避免編碼不同造成亂碼
rt = rate.text        # 以文字模式讀取內容
rts = rt.split('\n')     # 使用「換行」將內容拆分成串列
for i in rts:        # 讀取串列的每個項目
  try:            # 使用try 避開最後一行的空白行
    a = i.split(',')   # 每個項目用逗號拆分成子串列
    print(a[0] + ': ' + a[12])   # 取出第一個 ( 0 ) 和第十三個項目 ( 12 )
  except:
    break
```
